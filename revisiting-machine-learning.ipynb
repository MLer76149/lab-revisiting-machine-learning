- Check for null values in all the columns

```python
def show_nan(df):
    for col in df.columns:
        if df[col].isna().sum() > 0:
            print("The column", col, "has", df[col].isna().sum(), "NaN" )
    
```

```python
show_nan(categorical)
```

```python
def show_empty_variables(df):
    columns = df.columns
    for col in columns:
        if len(df[df[col].isin(["", " "])][col].value_counts()) == 1:
            empty_amount = df[df[col].isin(["", " "])][col].value_counts()[0]
            percent_empty = np.round((empty_amount / df.shape[0])*100, 2)
            print("The column", col, "has", empty_amount, "empty strings which means", percent_empty, "%" )
        
    
```

```python
show_empty_variables(categorical)
```

```python
categorical = categorical.apply(lambda x: x.replace(" ", np.NaN))
```

```python
show_nan(categorical)
```

- Identify columns that over 85% missing values

```python
def find_nan(df, threshold):
    nan_columns = pd.DataFrame(df.isna().sum()/len(df)).reset_index()
    #nan_columns.columns = ['column_name', 'nulls_percentage']
    nan_threshold_columns = nan_columns[nan_columns.iloc[:,1] > threshold]["index"]
    return list(nan_threshold_columns)
        
```

```python
nan_over_0_85 = find_nan(categorical, 0.85)
```

```python
display(nan_over_0_85)
```

- Remove those columns from the dataframe

```python
categorical = categorical.drop(columns=nan_over_0_85)
```

- Reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest

```python
categorical["GENDER"].value_counts(dropna=False)
```

```python
categorical["GENDER"] = categorical["GENDER"].apply(lambda x: "other" if (x in ["U", "J", "C", "A"]) else x)
```

```python
categorical['GENDER'] = categorical['GENDER'].fillna('F')
```
